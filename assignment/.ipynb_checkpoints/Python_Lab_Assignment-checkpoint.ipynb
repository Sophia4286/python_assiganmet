### Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#  Sort and display the top five schools in overall passing rate

pass_count =(school_data_complete[school_data_complete['math_score'] > 70].groupby(['school_name'],as_index=False)['Student ID'].count())
school_count=(school_data_complete.groupby(['school_name'], as_index=False)['Student ID'].count())

school_count['pass_rate']= pass_count['Student ID']/school_count['Student ID']

school_count.sort_values(by =['pass_rate'],ascending= False )[['school_name','pass_rate']].head(5)
```

```python
# Calculate total school budget 

school_unique= school_data_complete.groupby(by = ['school_name','budget'],as_index=False).count()[['school_name','budget']]
school_unique['budget'].sum()
```

```python
# Calculate per student budget

school_data
student_count=(school_data_complete.groupby(['School ID'], as_index=False)['Student ID'].count())

student_count

school_data['avg_budget'] = school_data['budget']/student_count['Student ID']
school_data
```

```python
# Cacluate the avg math and reading score
avg_math_score=school_data_complete.groupby(['School ID'], as_index=False)['math_score'].mean()
school_data['avg_math_score'] = avg_math_score['math_score']
school_data
```

#### Find the passing rate for math and reading (above 70 points)

```python
# Find the total counts of math result
school_count=(school_data_complete.groupby(['School ID'], as_index=False)['math_score'].count())
# Find the counts for math result in each school that pass 70 or higher
pass_count =(school_data_complete[school_data_complete['math_score'] > 70].groupby(['School ID'],as_index=False)['math_score'].count())
# Calculate the math passing rate
school_data['math_passing_rate'] = pass_count['math_score']/school_count['math_score']
school_data
```

```python
# Find the total counts of read result
school_count=(school_data_complete.groupby(['School ID'], as_index=False)['reading_score'].count())

# Find the counts for read result in each school that pass 70 or higher
pass_count =(school_data_complete[school_data_complete['reading_score'] > 70].groupby(['School ID'],as_index=False)['reading_score'].count())

# Calculate the read passing rate
school_data['reading_passing_rate'] = pass_count['reading_score']/school_count['reading_score']

school_data
```

```python
# Calculate the overall passing rate (average of the math and reading passing rate)

school_data['passing_rate'] = (school_data['math_passing_rate'] + school_data['reading_passing_rate'])/2
school_data

```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Create a new column to show budget per student in each row

results=pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
results=results.drop(['School ID','type','size','budget', 'avg_math_score', 'math_passing_rate','reading_passing_rate','passing_rate'], axis=1)
results
student_data = results
```

```python
# Create a new column to define the spending ranges per student
student_data['spending_ranges']=pd.cut(student_data.avg_budget,spending_bins, labels=group_names)
```

```python
# Calculate the average math score within each spending range


student_data.groupby(['spending_ranges'], as_index=False)['math_score'].mean()
```

```python
# Calculate the percentage passing rate for math in each spending range

range_pass_count =student_data[student_data['math_score']>70].groupby(['spending_ranges'], as_index=False)['Student ID'].count()
range_count=student_data.groupby(['spending_ranges'], as_index=False)['Student ID'].count()

range_count['math_passing_rate']=range_pass_count['Student ID']/range_count['Student ID']
range_count
```

```python
# Calculate the percentage overall passing rate in each spending range
range_count['passing_rate'] = (range_count['reading_passing_rate'] + range_count['math_passing_rate'])/2

range_count
```
